```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
soft_drink = pd.read_csv('Soft_drink.csv')
```

```python
soft_drink.Frequency
```

```python
frequency = np.array(soft_drink.Frequency)
```

```python
total = frequency.sum()
```

```python
total
```

```python
relative_frequency = np.array(frequency/total)
```

```python
percent_frequency = np.array(relative_frequency*100)
percent_frequency
```

```python
%matplotlib inline
colour = ["blue","green","red","yellow","orange"]
plt.bar(soft_drink.Soft_Drinks,frequency,width = 0.5,color = colour)
plt.xlabel("Drink")
plt.ylabel("Frequency")
plt.title("Bar Graph for Frequency")
plt.show
```

```python
plt.plot(soft_drink.Soft_Drinks,relative_frequency)  
plt.xlabel("Drink")
plt.ylabel("Relative_Frequency")
plt.title("Line Graph for Relative_Frequency")
plt.show
```

```python
plt.pie(percent_frequency,explode = [.1,.1,.1,.1,.1],labels = soft_drink.Soft_Drinks)
plt.title("Pie Chart for Percent_Frequency")
plt.show
```

```python
plt.ylabel('Frequency') 
plt.xlabel('Drink') 
plt.title("Scatter Plot for Frequency")
plt.scatter(soft_drink.Soft_Drinks,frequency)
plt.show()
```

```python
plt.title("Histogram for Frequency")
plt.hist(frequency)
plt.show()
```
